```python
import pandas as pd
import numpy as np
import tweepy
from dotenv import load_dotenv
import os
```

```python
# Load the environment variables from the .env file
load_dotenv()

# Get the API key, API secret key, access token, and access token secret from the environment variables
API_KEY = os.getenv('consumer_key')
API_SECRET_KEY = os.getenv('consumer_secret')
ACCESS_TOKEN = os.getenv('access_token')
ACCESS_TOKEN_SECRET = os.getenv('access_token_secret')
```

```python
# Create the API object
auth = tweepy.OAuthHandler(API_KEY, API_SECRET_KEY)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
api = tweepy.API(auth)
```

```python
users = ['BDMOFA', 'DrSJaishankar','MOFA_Taiwan','CanadaFP','StateDept','StateDRL','IRIMFA_EN','KSAmofaEN']
```

```python
def scrape_tweets(username, words):
    # Get all tweets from the user's timeline
    tweets = api.user_timeline(screen_name=username, count=50, tweet_mode='extended')
    # Create a DataFrame to store the tweets
    df = pd.DataFrame(columns=['user', 'text', 'created_at'])
    for tweet in tweets:
        # Check if the tweet contains the specific word
        if any(word in tweet.full_text for word in words):
            # Add the tweet to the DataFrame
            df = df.append({'user': tweet.user.screen_name, 'text': tweet.full_text, 'created_at': tweet.created_at}, ignore_index=True)
    return df
```

```python
words = ['Nato', 'Ukraine', 'Russia', 'Oil','Gas','Energy','Nuclear','Nuclear Energy','Nuclear Power','Chinese','Iran']
tweets_df = pd.DataFrame(columns=['user', 'text', 'created_at'])
for user in users:
    user_tweets = scrape_tweets(user, words)
    tweets_df = pd.concat([tweets_df, user_tweets])
```

```python
tweets_df.to_csv('tweets.csv', index=False)
```

```python
from datetime import datetime, timedelta
# Define search query and time range
query = "Bangladesh"
today = datetime.today()
since_date = today - timedelta(hours=1)

# Search for tweets containing query since since_date
tweets = tweepy.Cursor(api.search_tweets, q=query, lang="en", since_id=since_date).items()

# Loop through tweets and print them
for tweet in tweets:
    print(tweet.text)
```
